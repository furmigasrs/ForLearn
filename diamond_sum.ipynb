# DIAMOND SUM

```python
def diamond_sum(n):
    matriz = []
    di_sum, termo = 0,0
    diamante = []
    for l in range(n):
        linha = []
        for c in range(n):
            termo +=1
            linha.append(termo)
        matriz.append(linha)
        if n == 1:
            return 1
    for i in range(n):
        for r in range(int((n+1)/2)):
            if r == abs(int(((n-1)/2)-i)):
                diamante.append(matriz[r][i])
                if r != int((n-1)/2) and r != 0:
                    diamante.append(matriz[r+int((n-1)/2)][i])
            if r == int((n-1)/2) and i == int((n-1)/2):
                diamante.append(matriz[r+int((n-1)/2)][i])  
    return sum(diamante)

```
