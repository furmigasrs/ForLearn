# Rectangle Collision Detect

```python
class Rectangle:
    def __init__(self, x, y, w, h):       
        self.x = x
        self.y = y
        self.w = w
        self.h = h 


def intersecting(a, b):
    collide = False
    recax = [i for i in range(a.x,a.x+a.w+1)]
    recay = [i for i in range(a.y,a.y+a.h+1)]
    recbx = [i for i in range(b.x,b.x+b.w+1)]
    recby = [i for i in range(b.y,b.y+b.h+1)]
    for i in recax:
        if i in recbx:
            for i in recay:
                if i in recby:
                    return True
    return False
```

```python
a = Rectangle(10, 20, 100, 20)
b = Rectangle(10, 40, 15, 20)
c = Rectangle(50, 50, 20, 30)
d = Rectangle(90, 25, 100, 5)
```
